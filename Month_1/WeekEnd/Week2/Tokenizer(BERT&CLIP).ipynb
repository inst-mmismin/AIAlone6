```python
from transformers import BertTokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-multilingual-cased')
```

```python
text1 = '비가 주륵주륵 내리고 있습니다.'
text2 = 'It\'s raining hard and fast.'
```

```python
tokenized_text1 = tokenizer.tokenize(text1)
print(tokenized_text1)

tokenized_text2 = tokenizer.tokenize(text2)
print(tokenized_text2)
```

```python
text3 = 'Handle all the shared methods for tokenization and special tokens as well as methods'
tokenizer.tokenize(text3)
```

# CLIP tokenizer

```python
!pip install git+https://github.com/openai/CLIP.git
```

```python
# 참고 링크 : https://github.com/openai/CLIP/blob/a9b1bf5920416aaeaec965c25dd9e8f98c864f16/clip/simple_tokenizer.py#L121
from clip.simple_tokenizer import SimpleTokenizer
st = SimpleTokenizer()
```

```python
text1 = '비가 주륵주륵 내리고 있습니다.'
text2 = 'It\'s raining hard and fast.'
```

```python
def tokenize(text):
    encoded = st.encode(text)
    return [st.decoder[en] for en in encoded]
```

```python
tokenized1 = tokenize(text1)
print(tokenized1)

tokenized2 = tokenize(text2)
print(tokenized2)
```

```python
# BERT : ['It', "'", 's', 'rain', '##ing', 'hard', 'and', 'fast', '.']
# Clip : ['it', "'s", 'raining', 'hard', 'and', 'fast', '.']
```
